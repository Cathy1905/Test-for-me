NARX 1: Pilot Simulation, Data Generation, and Data Visualization

```python
def simulate_narx1(N: int = 1000, sigma: tuple = (0.1, 0.1), seed: int = 42):
    """Simulate pilot NARX1 with one input and two outputs."""
    np.random.seed(seed)
    u = np.random.uniform(-1, 1, size=(N, 1))   # Initialize inputs with random values
    u[0,0] = 0                                  # Initialize the first input to zero

    y = np.zeros((N, 2))                        # Initialize outputs to zero

    w1 = np.random.normal(0, sigma[0], size=(N, 1))
    w2 = np.random.normal(0, sigma[1], size=(N, 1))

    for k in range(1, N - 1):
        y[k + 1, 0] = 0.5 * y[k - 1, 1] + np.sin(y[k, 1]) + 0.3 * u[k - 1, 0] + w1[k,0]
        y[k + 1, 1] = 0.5 * y[k - 1, 0] + np.sin(y[k, 0]) + 0.2 * u[k, 0] + w2[k,0] #attention here, I changed from k, 0

    return u, y
```

```python
#Pilot Simulation
def simulate_narx2(N: int = 1000, sigma: tuple = (0.1, 0.1), seed: int = 42):
    """Simulate pilot NARX2 with two inputs and two outputs."""
    np.random.seed(seed)
    u = np.random.uniform(-1, 1, size=(N, 2))       # Initialize inputs with random values
    u[0, :] = 0                                     # Initialize the first input to zero

    y = np.zeros((N, 2))                            # Initialize outputs to zero

    w1 = np.random.normal(0, sigma[0], size=(N, 1)) # Initialize noise for output 1
    w2 = np.random.normal(0, sigma[1], size=(N, 1)) # Initialize noise for output 2

    for k in range(2, N - 1):
        # y1 computation
        y1_num = y[k, 0] * y[k - 1, 0] * y[k - 2, 0] * (y[k - 2, 0] - 1) * u[k - 1, 1] + u[k, 1]
        y1_den = 1 + y[k - 1, 1] ** 2 + y[k - 2, 1] ** 2
        y[k + 1, 0] = y1_num / y1_den + w1[k, 0]

        # y2 computation    
        y2_num = y[k, 1] * y[k - 1, 1] * y[k - 2, 1] * (y[k - 2, 1] - 1) * u[k - 1, 0] + u[k, 0]
        y2_den = 1 + y[k - 1, 0] ** 2 + y[k - 2, 0] ** 2
        y[k + 1, 1] = y2_num / y2_den + w2[k, 0]

        # Clip the outputs to prevent overflow in scalar multiply
        y[k + 1, :] = np.clip(y[k + 1, :], -2.5, 2.5)

    return u, y

```

```python
def rmse(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    """
    Root Mean Squared Error (multi-output).
    Correspond exactement à :
        sqrt( 1/(2N) * sum_{k,j} (y_j(k) - yhat_j(k))^2 )
    """
    return np.sqrt(np.mean((y_true - y_pred) ** 2))


def nmse(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    """
    Normalized Mean Squared Error (multi-output).
    Correspond exactement à :
        sum_{k,j} (y_j(k) - yhat_j(k))^2
        --------------------------------
        sum_{k,j} (y_j(k) - mean(y_j))^2
    """
    mean = np.mean(y_true, axis=0, keepdims=True)
    numerator = np.sum((y_true - y_pred) ** 2)
    denominator = np.sum((y_true - mean) ** 2)
    return numerator / denominator


def build_narx_features(u, y, na, nb, d):
    """
    Transforme les données en format supervisé.
    y(k+1) = f(y(k)...y(k-na+1), u(k-d)...u(k-d-nb+1))
    """
    N = len(y)
    X = []
    Y = []

    for k in range(N - 1):
        line = []
        
        # 1. Lags de la sortie Y : y(k), y(k-1), ..., y(k-na+1)
        for i in range(na):
            if k - i >= 0:
                line.extend(y[k - i]) # Ajoute [y1, y2]
            else:
                line.extend([0.0, 0.0]) # Condition initiale nulle
                
        # 2. Lags de l'entrée U avec délai d : u(k-d), ..., u(k-d-nb+1)
        for j in range(nb):
            idx_u = k - d - j
            if idx_u >= 0:
                line.extend(u[idx_u]) # Ajoute u1 (et u2 si MIMO)
            else:
                line.extend([0.0] * u.shape[1])
        
        X.append(line)
        Y.append(y[k + 1])
        
    return np.array(X), np.array(Y)
```

## 2B:  MIMO and multi-step-aheadstrategyselectionapproach: the main challenges of the project are i) the MIMO nature of the task and 2) that the prediction is expected not for a short horizon but for a very long one.                   

Challenge I: The MIMO Nature of NARX
The project involves a MIMO task with two outputs ($y_1, y_2$) and two inputs ($u_1, u_2$). To handle this simultaneously, our model does not treat the outputs as independent tasks. Instead, we use a vector-valued regression approach:Input Vector ($\mathcal{X}$): We construct a single feature vector that concatenates the lags of all variables:we implemented a global MIMO NARX architecture where the lags of all variables are fed into the same model. $$\mathcal{X}_k = [y_1(k-d \dots -n_a), y_2(k-d \dots -n_a), u_1(k \dots -n_b), u_2(k \dots -n_b)]$$ 

Output Vector ($\mathcal{Y}$): The model is trained to map $\mathcal{X}_k$ to the vector $\hat{Y}_{k+1} = [\hat{y}_1, \hat{y}_2]$.Simultaneous Learning: By using a regressor that supports multi-output targets (like Ridge or Random Forest), the model optimizes the prediction for both axes at the same time, preserving the physical "link" between them. The main challenge is that in a 1000-step recursive simulation, any error in one output dimension propagates to the others, significantly increasing the risk of error explosion and divergence that will be cover just below.

```python
def build_mimo_features(U, Y, na, nb, d):
    """
    Constructs the MIMO NARX feature matrix X and target matrix Y.
    Follows the structure defined in Equations (1) and (2) of the project.
    """
    n_samples = len(Y)
    max_lag = max(na + d, nb)
    X, target = [], []

    for k in range(max_lag, n_samples):
        # 1. Output Lags: Collecting past of BOTH y1 and y2
        y_lags = []
        for i in range(d, na + d):
            # Y[k-i] contains [y1, y2] simultaneously
            y_lags.extend(Y[k - i - 1]) 
            
        # 2. Input Lags: Collecting past of BOTH u1 and u2
        u_lags = []
        for j in range(nb + 1):
            # U[k-j] contains [u1, u2] simultaneously
            u_lags.extend(U[k - j])
            
        # Concatenate into one MIMO feature row
        X.append(np.hstack([y_lags, u_lags]))
        # Target is the vector [y1, y2] at the next step
        target.append(Y[k])

    return np.array(X), np.array(target)

# --- Training the MIMO Model ---
# Utr and Ytr have shapes (1000, 2) as per section 2.2
X_train, Y1_train = build_mimo_features(U_train, Y_train, na=5, nb=10, d=0)

# The regressor learns f1 and f2 simultaneously
mimo_model = Ridge(alpha=1.0)
mimo_model.fit(X_train, Y1_train) 

# mimo_model.predict(X) will now return an array of shape (N, 2)
# effectively providing [y_hat1, y_hat2] in one call.
```

Error Propagation in Multi-step Forecasting

The transition from a one-step-ahead validation ($H=1$) to a long-term recursive simulation ($H=1000$) introduces a major statistical challenge. In recursive mode, each predicted value $\hat{y}_{t+1}$ is reused as an input feature to predict the subsequent step $y_{t+2}$.From a statistical perspective, this process leads to the systematic accumulation of prediction errors. As demonstrated by Ben Taieb (2012) and Bontempi, even a negligible bias at the start of the simulation can grow exponentially, leading to a total divergence of the trajectory. This phenomenon is particularly critical for high-variance models which, by over-fitting to local noise, lose global stability during autonomous simulation.

2. Theoretical Framework: The Bias-Variance Trade-off

The architecture of our forecasting system is grounded in a rigorous analysis of the bias-variance trade-off:

Linear Models (High Bias, Low Variance): While they may lack the flexibility to capture complex non-linear dynamics, they offer structural stability. Their low variance prevents the simulation from diverging toward nonsensical or infinite values.

Machine Learning Models (Low Bias, High Variance): Complex algorithms such as Random Forests or Gradient Boosting capture non-linearities effectively but are "volatile." When used in a purely recursive mode, their instability propagates at every iteration.

3. Proposed Solution: The "Boost" Strategy
To overcome these limitations, we implemented the "Boost Strategy" (a multi-stage forecasting strategy) proposed by Ben Taieb. Rather than treating linear models and Machine Learning as competitors, this approach leverages them as complementary components through a two-step process:

The Linear Foundation (Global Stability): A linear NARX (or Ridge) model defines the baseline trajectory. It ensures that the simulation remains within credible physical boundaries.

Gradient Boosting Rectification (Local Precision): We train a Gradient Boosting algorithm specifically on the residuals (the errors) of the linear model.

4. Statistical Justification of the Hybrid Approach
This hybrid methodology is superior for several reasons identified in the SML literature:

Mitigation of Divergence Risk: The boosting model does not attempt to predict the pilot's position directly, but rather the prediction error of the linear model. Since this error is a signal of low magnitude, the impact of any boosting error on the global simulation remains minimal.

Model Synergy: We benefit from the low variance of the linear model for temporal structure and the low bias of Gradient Boosting for the precision of non-linear maneuvers.


3. Methodological Alternatives and Selection

To prevent divergence, we evaluated three distinct strategies.

Strategy A, the “recursive pure ML” approach, uses a high-capacity model such as a Random Forest. The observation is that, while it is very accurate in the short term, its high variance makes it unstable. It tends to overfit local noise, which causes the recursive loop to drift and diverge rapidly.

Strategy B, the “recursive pure linear” approach, relies on a standard Linear or Ridge regression. This method is extremely stable due to its low variance. However, it suffers from high structural bias: it “smooths out” the pilot’s non-linear maneuvers, resulting in poor accuracy.

Finally, Strategy C, the “hybrid Boost strategy,” based on the work of Ben Taieb and Bontempi (2012), was selected. This multi-stage approach uses a linear model as a stable “skeleton” to anchor the global trajectory (low variance), while a Gradient Boosting model acts as a “rectifier” to capture non-linear errors (low bias).


```python
from sklearn.linear_model import Ridge

def simulate_linear(u_tr, y_tr, u_ts, na, nb, d):
    X_tr, Y_tr = build_narx_dataset(u_tr, y_tr, na, nb, d)
    model = Ridge().fit(X_tr, Y_tr) # Modèle stable (Low Variance)
    
    y_hist = list(y_tr[-na:])
    preds = []
    for k in range(1000):
        # Construction du vecteur MIMO (Lags de y1, y2, u1, u2)
        feat = build_feature_vector(y_hist, u_ts, k, na, nb, d) 
        y_next = model.predict(feat.reshape(1, -1))[0]
        preds.append(y_next)
        y_hist.append(y_next) # Réinjection (Recursive)
    return np.array(preds)
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor

def simulate_hybrid(u_tr, y_tr, u_ts, na, nb, d):
    X_tr, Y_tr = build_narx_dataset(u_tr, y_tr, na, nb, d)
    # 1. Base Linéaire
    model_lin = Ridge().fit(X_tr, Y_tr)
    # 2. Rectificateur (Boosting sur les erreurs du linéaire)
    residus = Y_tr - model_lin.predict(X_tr)
    model_err = HistGradientBoostingRegressor().fit(X_tr, residus.ravel())
    
    y_hist = list(y_tr[-na:])
    preds = []
    for k in range(1000):
        feat = build_feature_vector(y_hist, u_ts, k, na, nb, d).reshape(1, -1)
        # Prédiction = Tendance stable + Correction non-linéaire
        y_next = model_lin.predict(feat)[0] + model_err.predict(feat)[0]
        preds.append(y_next)
        y_hist.append(y_next)
    return np.array(preds)
```

```python
"""
NARX Multi-Strategy Benchmark - CORRECTED VERSION
==================================================
Uses OPTIMAL parameters for each pilot (different na, nb, d per pilot)

Author: Corrected to use best parameters per pilot (e.g., NARX1: 5,10,0 / NARX2: 2,3,0)
"""

import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor


def build_narx_dataset(U, Y, na, nb, d):
    """
    Builds MIMO NARX dataset with consistent feature construction.
    
    Args:
        U: Input array (N, 2) - [u1, u2]
        Y: Output array (N, 2) - [y1, y2]
        na: Number of output lags
        nb: Number of input lags (actual count is nb+1, from 0 to nb)
        d: Delay
    
    Returns:
        X: Feature matrix (N', na*2 + (nb+1)*2)
        Y_target: Target matrix (N', 2)
    """
    n_samples = len(Y)
    max_lag = max(na + d, nb + d)
    
    X_list, Y_list = [], []
    
    for k in range(max_lag, n_samples):
        # Output lags: Y[k-d-1], Y[k-d-2], ..., Y[k-d-na]
        y_lags = []
        for i in range(1, na + 1):
            y_lags.extend(Y[k - d - i])
        
        # Input lags: U[k-d], U[k-d-1], ..., U[k-d-nb]
        u_lags = []
        for j in range(nb + 1):
            u_lags.extend(U[k - d - j])
        
        X_list.append(np.hstack([y_lags, u_lags]))
        Y_list.append(Y[k])
    
    return np.array(X_list), np.array(Y_list)


def build_feature_vector_consistent(y_hist, u_ts, k, na, nb, d):
    """
    Builds a single feature vector during simulation.
    CRITICAL: Must exactly match build_narx_dataset logic!
    """
    feat = []
    
    # Output lags
    for i in range(1, na + 1):
        if len(y_hist) >= i:
            feat.extend(y_hist[-i])
        else:
            feat.extend([0.0, 0.0])
    
    # Input lags
    for j in range(nb + 1):
        idx = k - d - j
        if 0 <= idx < len(u_ts):
            feat.extend(u_ts[idx])
        else:
            feat.extend([0.0, 0.0])
    
    return np.array(feat)


def run_simulation_fixed(model_type, X_tr, Y_tr, u_ts, y_tr, na, nb, d):
    """
    Performs 1000-step recursive MIMO NARX simulation.
    """
    # === TRAINING PHASE ===
    if model_type == 'Linear':
        model = Ridge(alpha=1e-6).fit(X_tr, Y_tr)
        
    elif model_type == 'Pure_ML':
        model = RandomForestRegressor(
            n_estimators=50, 
            max_depth=10, 
            random_state=42
        ).fit(X_tr, Y_tr)
        
    elif model_type == 'Hybrid':
        model_lin = Ridge(alpha=1e-6).fit(X_tr, Y_tr)
        residuals = Y_tr - model_lin.predict(X_tr)
        model_err1 = HistGradientBoostingRegressor(random_state=42).fit(X_tr, residuals[:, 0])
        model_err2 = HistGradientBoostingRegressor(random_state=42).fit(X_tr, residuals[:, 1])
    
    # === SIMULATION PHASE ===
    y_hist = list(y_tr[-na:])
    predictions = []
    
    for k in range(1000):
        feat = build_feature_vector_consistent(y_hist, u_ts, k, na, nb, d)
        X_in = feat.reshape(1, -1)
        
        if model_type == 'Hybrid':
            y_base = model_lin.predict(X_in)[0]
            y_next = np.array([
                y_base[0] + model_err1.predict(X_in)[0],
                y_base[1] + model_err2.predict(X_in)[0]
            ])
        else:
            y_next = model.predict(X_in)[0]
        
        predictions.append(y_next)
        y_hist.append(y_next)
    
    return np.array(predictions)


def benchmark_pilots_with_optimal_params(pilot_configs):
    """
    Benchmarks strategies using OPTIMAL parameters for each pilot.
    
    Args:
        pilot_configs: List of dicts, each containing:
            - 'name': Pilot name (e.g., "NARX1", "NARX2")
            - 'data': Dict with "Utr", "Ytr", "Uts", "Yts"
            - 'na': Optimal na for this pilot
            - 'nb': Optimal nb for this pilot
            - 'd': Optimal d for this pilot
    
    Returns:
        DataFrame with benchmark results
    """
    results = []
    
    for config in pilot_configs:
        pilot_name = config['name']
        pilot_data = config['data']
        na, nb, d = config['na'], config['nb'], config['d']
        
        print(f"\n{'='*70}")
        print(f"{pilot_name} - Optimal params: na={na}, nb={nb}, d={d}")
        print(f"{'='*70}")
        
        # Build training dataset with OPTIMAL parameters
        X_tr, Y_tr = build_narx_dataset(
            pilot_data["Utr"], pilot_data["Ytr"], na, nb, d
        )
        y_true = pilot_data["Yts"][:1000]
        
        expected_features = na * 2 + (nb + 1) * 2
        print(f"Training: {X_tr.shape[0]} samples, {X_tr.shape[1]} features")
        print(f"Expected: {expected_features} features")
        
        # Verify feature consistency
        if X_tr.shape[1] != expected_features:
            print(f"⚠️  WARNING: Feature mismatch!")
            continue
        
        # Test each strategy
        for strategy in ['Linear', 'Pure_ML', 'Hybrid']:
            print(f"\n  {strategy}...", end=" ", flush=True)
            
            try:
                y_sim = run_simulation_fixed(
                    strategy, X_tr, Y_tr, pilot_data["Uts"],
                    pilot_data["Ytr"], na, nb, d
                )
                
                # Stability check
                max_true = np.max(np.abs(y_true))
                threshold = max_true * 20
                is_stable = np.all(np.abs(y_sim) < threshold)
                
                if is_stable:
                    rmse = np.sqrt(np.mean((y_true - y_sim)**2))
                    y_mean = np.mean(y_true, axis=0)
                    ss_tot = np.sum((y_true - y_mean)**2)
                    ss_res = np.sum((y_true - y_sim)**2)
                    nmse = ss_res / ss_tot if ss_tot > 0 else np.inf
                    status = "Stable"
                    print(f"✅ RMSE={rmse:.4f}, NMSE={nmse:.4f}")
                else:
                    rmse, nmse = np.inf, np.inf
                    status = "EXPLOSION"
                    max_pred = np.max(np.abs(y_sim))
                    print(f"❌ EXPLOSION (max={max_pred:.2f})")
                
                results.append({
                    'Pilot': pilot_name,
                    'na': na, 'nb': nb, 'd': d,
                    'Strategy': strategy,
                    'RMSE': rmse,
                    'NMSE': nmse,
                    'Status': status
                })
                
            except Exception as e:
                print(f"❌ ERROR: {e}")
                results.append({
                    'Pilot': pilot_name,
                    'na': na, 'nb': nb, 'd': d,
                    'Strategy': strategy,
                    'RMSE': np.nan,
                    'NMSE': np.nan,
                    'Status': f"Error"
                })
    
    return pd.DataFrame(results)


# ============================================================================
# USAGE EXAMPLE
# ============================================================================

if __name__ == "__main__":
    
    # Demo data (replace with your actual pilot1 and pilot2)
    np.random.seed(42)
    pilot1_data = {
        "Utr": np.random.randn(1000, 2),
        "Ytr": np.random.randn(1000, 2),
        "Uts": np.random.randn(1500, 2),
        "Yts": np.random.randn(1500, 2)
    }
    pilot2_data = {
        "Utr": np.random.randn(1000, 2),
        "Ytr": np.random.randn(1000, 2),
        "Uts": np.random.randn(1500, 2),
        "Yts": np.random.randn(1500, 2)
    }
    
    # ========== CONFIGURATION AVEC PARAMÈTRES OPTIMAUX ==========
    # IMPORTANT: Utilisez les paramètres trouvés par votre grid search!
    pilot_configs = [
        {
            'name': 'NARX1',
            'data': pilot1_data,
            'na': 5,   # ← Paramètres optimaux pour NARX1
            'nb': 10,  # ← (trouvés par grid search)
            'd': 0
        },
        {
            'name': 'NARX2',
            'data': pilot2_data,
            'na': 2,   # ← Paramètres optimaux pour NARX2
            'nb': 3,   # ← (trouvés par grid search)
            'd': 0
        }
    ]
    
    print("="*70)
    print("BENCHMARK AVEC PARAMÈTRES OPTIMAUX PAR PILOT")
    print("="*70)
    
    # Run benchmark
    df_results = benchmark_pilots_with_optimal_params(pilot_configs)
    
    # Display results
    print("\n" + "="*70)
    print("RÉSULTATS FINAUX")
    print("="*70)
    print(df_results.to_string(index=False))
    
    # Best per pilot
    print("\n" + "="*70)
    print("MEILLEURE STRATÉGIE PAR PILOT (NMSE)")
    print("="*70)
    
    for pilot_name in df_results['Pilot'].unique():
        pilot_results = df_results[df_results['Pilot'] == pilot_name]
        stable_results = pilot_results[pilot_results['Status'] == 'Stable']
        
        if len(stable_results) > 0:
            best = stable_results.loc[stable_results['NMSE'].idxmin()]
            print(f"\n{pilot_name} (na={best['na']}, nb={best['nb']}, d={best['d']}):")
            print(f"  Meilleure stratégie: {best['Strategy']}")
            print(f"  RMSE: {best['RMSE']:.6f}")
            print(f"  NMSE: {best['NMSE']:.6f}")
        else:
            print(f"\n{pilot_name}: Tous les modèles ont explosé!")
    
    # Comparison table
    print("\n" + "="*70)
    print("COMPARAISON PAR STRATÉGIE")
    print("="*70)
    pivot = df_results[df_results['Status'] == 'Stable'].pivot_table(
        index='Strategy', 
        columns='Pilot', 
        values='NMSE'
    )
    print(pivot)
```
